# Fine-tune YOLO26-s on the NEON tree-crown benchmark (cloud GPU)

A thin wrapper around [`scripts/train_yolo.py`](https://github.com/tarpous/urban-canopy-detection/blob/main/scripts/train_yolo.py). Runs on a Colab/Kaggle GPU for anyone without a local card; the local, primary path is the same script on your own machine. Metrics land in `results/metrics.json` — commit that back to update the README table via `canopy make-table`.

```python
# Thin wrapper: clone the repo and install the training extras. All the real
# logic lives in the tested `urban_canopy` package and `scripts/train_*.py`;
# this notebook only runs them on a cloud GPU for anyone without a local one.
import subprocess, sys, os
if not os.path.exists("urban-canopy-detection"):
    subprocess.run(["git", "clone", "--depth", "1",
                    "https://github.com/tarpous/urban-canopy-detection.git"], check=True)
%cd urban-canopy-detection
subprocess.run([sys.executable, "-m", "pip", "-q", "install", "-e", "."], check=True)
```

```python
subprocess.run([sys.executable, '-m', 'pip', '-q', 'install', 'ultralytics'], check=True)
```

```python
# Fetch the NEON evaluation split (RGB + annotations). ~3.9 GB; the split, tiling
# and scoring are identical to the local run (site-disjoint, seed 0).
subprocess.run([sys.executable, "scripts/download_neon.py", "--evaluation", "--annotations"],
               check=True)
```

```python
subprocess.run([sys.executable, 'scripts/train_yolo.py',
                '--model', 'yolo26s.pt', '--epochs', '80', '--device', '0'], check=True)
```

```python
subprocess.run([sys.executable, 'scripts/run_baseline.py'], check=True)  # DeepForest, same val tiles
subprocess.run([sys.executable, 'scripts/make_results_table.py'], check=True)
```
